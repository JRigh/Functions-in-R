# Stirling's approximation

```python
import math

def stirling(n):
    return math.sqrt(2 * math.pi * n) * (n**n) * math.exp(-n)

# Comparison
approximation = stirling(15)
exact = math.factorial(15)

print("Approximation:", approximation)
print("Exact value:", exact)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define Stirling function
def stirling(n):
    return np.sqrt(2 * np.pi * n) * (n**n) * np.exp(-n)

# Generate data
index = np.arange(1, 4, 0.01)
exact = np.array([math.factorial(int(i)) for i in index])
stirlings = stirling(index)

# Plot
plt.plot(index, exact, label='Exact factorial', color='orange')
plt.plot(index, stirlings, label='Stirling approximation', color='green')
plt.title('Exact factorial vs Stirling approximation')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
